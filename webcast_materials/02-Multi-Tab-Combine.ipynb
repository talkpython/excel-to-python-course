### Worksheets with multiple tabs

Pandas can read Excel worksheets with multiple tabs

```python
import pandas as pd
from pathlib import Path
```

```python
excel_file = Path.cwd() / 'data' / 'raw' / '2019_Sales_Total_Tabs.xlsx'
```

```python
all_dfs = pd.read_excel(excel_file, sheet_name=None)
```

```python
all_dfs.keys()
```

```python
# No need to copy and paste all those values
df = pd.concat(all_dfs, ignore_index=True)
```
